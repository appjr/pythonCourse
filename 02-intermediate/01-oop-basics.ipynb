## 📦 Classes and Objects


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return f"Hi, I'm {self.name}"

p = Person("Alice", 25)
print(p.greet())
```

## 🎓 Key Takeaways
✅ Classes define objects
✅ `__init__` is constructor
✅ `self` refers to instance
