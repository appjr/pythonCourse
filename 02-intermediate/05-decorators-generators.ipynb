# 🐍 Intermediate: Decorators & Generators

---


## ⭐ Decorators


```python
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

## 🔄 Generators


```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)
```

## 🎓 Key Takeaways
✅ Decorators modify functions
✅ `yield` creates generators
✅ Generators are memory efficient
