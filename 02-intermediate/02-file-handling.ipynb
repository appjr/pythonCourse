## 📄 Reading and Writing Files


```python
# Writing to a file
with open("test.txt", "w") as f:
    f.write("Hello, World!")

# Reading from a file
with open("test.txt", "r") as f:
    content = f.read()
    print(content)
```

## 📊 CSV Files


```python
import csv

# Writing CSV
with open("data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Alice", 25])
```

## 🎓 Key Takeaways
✅ Use `with` for file handling
✅ Modes: r, w, a, r+
✅ CSV module for structured data
