## 🔀 Threading


```python
import threading
import time

def worker(name):
    print(f"{name} starting")
    time.sleep(2)
    print(f"{name} done")

threads = []
for i in range(3):
    t = threading.Thread(target=worker, args=(f"Worker-{i}",))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

## 🎓 Key Takeaways
✅ Threading for I/O-bound tasks
✅ Multiprocessing for CPU-bound
✅ async/await for modern concurrency
