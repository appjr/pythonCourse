## âž• Arithmetic Operators

These operators perform mathematical operations.

```python
# Basic arithmetic
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

## ðŸ” Comparison Operators

Compare values and return True or False.

```python
x = 5
y = 10

print(f"Equal: {x} == {y} â†’ {x == y}")
print(f"Not equal: {x} != {y} â†’ {x != y}")
print(f"Greater: {x} > {y} â†’ {x > y}")
print(f"Less: {x} < {y} â†’ {x < y}")
print(f"Greater or equal: {x} >= {y} â†’ {x >= y}")
print(f"Less or equal: {x} <= {y} â†’ {x <= y}")
```

## ðŸ”— Logical Operators

Combine conditional statements.

```python
# and, or, not
age = 25
has_license = True

print(f"Can drive: {age >= 18 and has_license}")
print(f"Eligible: {age < 18 or has_license}")
print(f"Not has_license: {not has_license}")
```

## ðŸ“Œ Assignment Operators

```python
x = 10
print(f"Initial: x = {x}")

x += 5  # x = x + 5
print(f"After x += 5: {x}")

x -= 3  # x = x - 3
print(f"After x -= 3: {x}")

x *= 2  # x = x * 2
print(f"After x *= 2: {x}")

x //= 4  # x = x // 4
print(f"After x //= 4: {x}")
```
