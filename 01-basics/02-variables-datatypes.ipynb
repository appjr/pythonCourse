## ðŸ“¦ What are Variables?

A **variable** is like a labeled container that stores data. You can think of it as a box with a name on it.

### Creating Variables:

````python
# Creating variables
name = "Alice"
age = 25
height = 5.6
is_student = True

print(name)
print(age)
print(height)
print(is_student)
````

### Variable Naming Rules:

âœ… **Good Variable Names:**
```python
name = "John"
age = 30
user_email = "john@example.com"
firstName = "John"  # camelCase
first_name = "John"  # snake_case (Python convention)
```

âŒ **Invalid Variable Names:**
```python
2name = "John"      # Can't start with number
first-name = "John" # Can't use hyphens
first name = "John" # Can't have spaces
class = "Math"      # Can't use Python keywords
```

### Python Naming Conventions:
- Use lowercase with underscores: `user_age`, `total_price`
- Be descriptive: `count` is better than `c`
- Constants use ALL_CAPS: `MAX_SIZE = 100`

## ðŸ”¢ Numbers: Integers and Floats

Python has two main number types:
- **int** (integer): Whole numbers like 1, 100, -5
- **float** (floating point): Decimal numbers like 3.14, -0.5, 2.0

````python
# Integers
age = 25
temperature = -5
year = 2024

print(f"Age: {age}")
print(f"Temperature: {temperature}")
print(f"Year: {year}")
````

````python
# Floats
pi = 3.14159
price = 19.99
height = 5.8

print(f"Pi: {pi}")
print(f"Price: ${price}")
print(f"Height: {height} feet")
````

### Basic Math Operations:

````python
# Math with integers
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Integer Division: {a} // {b} = {a // b}")
print(f"Modulus (remainder): {a} % {b} = {a % b}")
print(f"Power: {a} ** {b} = {a ** b}")
````

### Complex Numbers (Advanced):

````python
# Complex numbers (for advanced math)
complex_num = 3 + 4j
print(f"Complex number: {complex_num}")
print(f"Real part: {complex_num.real}")
print(f"Imaginary part: {complex_num.imag}")
````

## ðŸ“ Strings

Strings are sequences of characters (text). They can be created with single quotes `'` or double quotes `"`.

````python
# Creating strings
name = "Alice"
message = 'Hello, World!'
quote = "She said, 'Hello!'"

print(name)
print(message)
print(quote)
````

### Multi-line Strings:

````python
# Multi-line strings with triple quotes
paragraph = """
This is a multi-line string.
It can span multiple lines.
Very useful for long text!
"""

print(paragraph)
````

### String Operations:

````python
# String concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name

print(full_name)
````

````python
# String repetition
laugh = "ha" * 5
print(laugh)  # Output: hahahahaha
````

````python
# String length
text = "Python"
print(f"The word '{text}' has {len(text)} characters")
````

### String Indexing and Slicing:

````python
# String indexing (starts at 0)
word = "Python"

print(f"First character: {word[0]}")
print(f"Second character: {word[1]}")
print(f"Last character: {word[-1]}")
print(f"Second to last: {word[-2]}")
````

````python
# String slicing
text = "Hello, World!"

print(f"First 5 characters: {text[0:5]}")
print(f"From index 7 to end: {text[7:]}")
print(f"Last 6 characters: {text[-6:]}")
print(f"Every other character: {text[::2]}")
````

### Useful String Methods:

````python
text = "  Hello, Python!  "

print(f"Original: '{text}'")
print(f"Uppercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"Title case: {text.title()}")
print(f"Strip whitespace: '{text.strip()}'")
print(f"Replace: {text.replace('Python', 'World')}")
````

````python
# Checking string content
email = "user@example.com"

print(f"Starts with 'user': {email.startswith('user')}")
print(f"Ends with '.com': {email.endswith('.com')}")
print(f"Contains '@': {'@' in email}")
````

````python
# Splitting and joining
sentence = "Python is awesome"
words = sentence.split()  # Split by spaces
print(f"Words: {words}")

joined = "-".join(words)
print(f"Joined: {joined}")
````

### String Formatting:

````python
name = "Alice"
age = 25
height = 5.6

# F-strings (modern and recommended)
print(f"My name is {name}, I'm {age} years old")

# Format with precision
print(f"Height: {height:.1f} feet")  # 1 decimal place

# .format() method (older way)
print("Name: {}, Age: {}".format(name, age))

# % formatting (very old, avoid)
print("Name: %s, Age: %d" % (name, age))
````

## âœ… Booleans

Booleans represent True or False values. They're essential for decision-making in programs.

````python
# Boolean values
is_student = True
is_employed = False
has_license = True

print(f"Is student: {is_student}")
print(f"Is employed: {is_employed}")
print(f"Has license: {has_license}")
````

````python
# Comparison operations return booleans
age = 18

print(f"Age is 18: {age == 18}")
print(f"Age is greater than 21: {age > 21}")
print(f"Age is less than 21: {age < 21}")
print(f"Age is not 21: {age != 21}")
````

````python
# Truthy and Falsy values
# These evaluate to False: False, None, 0, 0.0, "", [], {}, ()
# Everything else evaluates to True

print(f"bool(0): {bool(0)}")
print(f"bool(1): {bool(1)}")
print(f"bool(''): {bool('')}")
print(f"bool('hello'): {bool('hello')}")
````

## â“ None Type

`None` represents the absence of a value. It's like an empty placeholder.

````python
# None type
result = None

print(f"Result: {result}")
print(f"Type: {type(result)}")

# Checking for None
if result is None:
    print("Result is None!")
````

## ðŸ”„ Type Checking and Conversion

### Checking Types:

````python
# type() function
name = "Alice"
age = 25
height = 5.6
is_student = True

print(f"Type of name: {type(name)}")
print(f"Type of age: {type(age)}")
print(f"Type of height: {type(height)}")
print(f"Type of is_student: {type(is_student)}")
````

### Type Conversion:

````python
# String to Integer
age_str = "25"
age_int = int(age_str)
print(f"String to Int: {age_str} -> {age_int}")
print(f"Type: {type(age_int)}")
````

````python
# String to Float
price_str = "19.99"
price_float = float(price_str)
print(f"String to Float: {price_str} -> {price_float}")
print(f"Type: {type(price_float)}")
````

````python
# Number to String
age = 25
age_str = str(age)
print(f"Int to String: {age} -> '{age_str}'")
print(f"Type: {type(age_str)}")
````

````python
# Float to Integer (truncates decimal)
height = 5.9
height_int = int(height)
print(f"Float to Int: {height} -> {height_int}")
````

````python
# Common conversion error
try:
    num = int("hello")  # This will cause an error
except ValueError as e:
    print(f"Error: {e}")
    print("Can't convert 'hello' to an integer!")
````
