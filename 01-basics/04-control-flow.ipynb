## üîÄ If Statements

Make decisions in your code.

```python
age = 18

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

### If-elif-else Chains

```python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"Score: {score}, Grade: {grade}")
```

## üîÅ For Loops

Iterate over sequences.

```python
# Loop with range
for i in range(5):
    print(f"Number: {i}")
```

```python
# Loop through a list
fruits = ['apple', 'banana', 'orange']
for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# Loop with index
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

## ‚ö° While Loops

Loop while a condition is true.

```python
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1
```

## üõë Break, Continue, Pass

```python
# Break: exit loop
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue: skip iteration
for i in range(5):
    if i == 2:
        continue
    print(i)
```
