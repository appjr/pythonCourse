## 🔄 Recurrent Neural Networks

RNNs process sequential data.

**Applications:**
- Time series
- Text generation
- Speech recognition


```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.LSTM(64, input_shape=(10, 1)),
    keras.layers.Dense(1)
])

print(model.summary())
```

## 🎓 Key Takeaways
✅ RNNs for sequences
✅ LSTM handles long-term dependencies
✅ GRU is simpler alternative
